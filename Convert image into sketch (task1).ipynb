```python
import cv2
```

```python
path = r'C:\Users\ADMIN\Pictures\Sildeshow\vampire-diaries-02.jpg'
image = cv2.imread(path)  
grey_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) # cv2.cvtColor(): convert an image from one color space to another 
invert = cv2.bitwise_not(grey_img)  # helps in masking of the image
blur = cv2.GaussianBlur(invert, (21, 21), 0) #GaussianBlur(): sharp edges in images are smoothed while minimizing too much blurring 
invertedblur = cv2.bitwise_not(blur)
sketch = cv2.divide(grey_img, invertedblur, scale=256.0)
```

```python
cv2.imwrite("sketch.jpg", sketch)  # converted image is saved as mentioned name
```
